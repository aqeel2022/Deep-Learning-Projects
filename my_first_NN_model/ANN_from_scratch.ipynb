In this notebook we will learn to create neural  network from scratch.<br> Without using any python ML library. It will give us the understanding of the basic working principle of ANN

We will create NeuralNetwork class with the helper class to train the Neural Network. <br> We will use the sigmoid function as our activation function

```python
import numpy as np
```

```python
class NeuralNetwork():
    def __init__(self):
        #seed for random number generation
        np.random.seed(1)
        #converting the weights to a 3x3 matrix
        self.synaptic_weights =2*np.random.random((3,1))-1
    
    def sigmoid(self,x):
        return 1/(1+ np.exp(-x))
    
    def sigmoid_deriviative(self,x):
        return x*(1-x)
   
    def train(self, training_inputs, training_outputs, training_iterations):
        #training the model to make accurate predictions while adjusting the weights
        
        for iteration in range(training_iterations):
            output=self.think(training_outputs)
            
            error =training_outputs - output
            
            adjustments= np.dot(training_inputs.T, error * self.sigmoid_deriviative(output))
            self.synaptic_weights +=adjustments
            
    def think(self, inputs):
        inputs=inputs.astype(float)
        output=self.sigmoid(np.dot(inputs, self.synaptic_weights))
        return output
    
```
